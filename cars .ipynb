```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.ad.co.il/car?sp261=13914"
html = requests.get(url)
soup = BeautifulSoup(html.content,'html.parser')

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'} # This is chrome, you can set whatever browser you like

response = requests.get(url,headers= headers)
if response.status_code == 200:
    print("Success")
else:
    print("Failure")
```

```python
all = soup.find_all('div', {'class':'card-body p-md-3'})
links=[]
for r in all:
    links.append("https://www.ad.co.il" + r.find('a').get('href'))
    
```

```python
links=links[:-3] #קופצות 3 מודעות שהם לא קשורות לדייטהסו
```

```python
len(links)
```

```python
cars = []
for link in links:
    html = requests.get(link)
    soup = BeautifulSoup(html.content,'html.parser') 
    data = soup.find('table', {'class':'table table-sm mb-4'})
    tr=   data.find_all('tr') 
    data_car = {}
    
    for tr in soup.find_all('tr'):
        tds = tr.find_all('td')
        if len(tds) == 2:
            key = tds[0].get_text(strip=True)
            value = tds[1].get_text(strip=True)
            data_car[key] = value
    
    h2_elements =  soup.find_all('h2', {'class':'card-title'})
    title_text = h2_elements[0].text.strip()
    parts = title_text.split(' ', 1)  # Limit split to 1 occurrence
    data_car["model"] = parts[1]
    data_car["manufactor"] = parts[0]
    data_car["Price"] = h2_elements[1].text.strip()
    
    img =  soup.find_all('div', {'class':'justify-content-center px-1'})
    data_car['Pic_num']=len(img)
    
    div_elements =  soup.find_all('div', {'class':'px-3'})
    create_date = div_elements[0].text.strip()
    jump_date = div_elements[1].text.strip()
    
    data_car["Cre_date"] =create_date.split(':')[1].strip()
    data_car["Repub_date"] =jump_date.split(':')[1].strip()
    
    import re
    try:
        des = soup.find('p', {'class': 'text-word-break'}).text
        cleaned_text = re.sub(r'<[^>]*>', '', des)
        # Remove English characters and signs, keeping Hebrew and whitespace
        cleaned_text = re.sub(r'[a-zA-Z\d\-!@#$%^&*()+=<>/?\\|{}\[\]:;`~]', '', cleaned_text)
        cleaned_text = cleaned_text.replace('\n', ' ')
        data_car["Description"] = cleaned_text
    except AttributeError:  # Handle case where p element or text attribute is missing
        data_car["Description"] = ''  # Set an empty string as default

    cars.append(data_car)


        
```

```python
import pandas as pd
df=pd.DataFrame(cars)
```
